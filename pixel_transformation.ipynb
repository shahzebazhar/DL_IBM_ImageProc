```python
import cv2
from matplotlib import pyplot as plt
```

```python
#Histograms basically measure the frequency of each intensity level in an image.
#For greyscale images, this history has 256 bins, as there are total 256 intensity levels in the image

barbara = cv2.imread("barbara.png", cv2.IMREAD_GRAYSCALE)
hist = cv2.calcHist([barbara], [0], None, [256], [0, 255])
# [barbara] is the image. [0] is the channel(color channel) that we would like to calculate.
# [256] is the number of bins, and [0, 255] is the range of intensity values (min, max)

plt.plot(hist)
```

```python
#Intensity transformation changes an image one pixel at a time.
#Basically, intensity transformation function converts one pixel and only bothers with that pixel at that time. It does not
#alter any neighboring pixels.
```

```python
#Image negative is where we reverse the intensity level of an image
#This also basically is a transform function, and is applied as 255 - I[i, j] where I is the image.
plt.imshow(cv2.cvtColor(barbara, cv2.COLOR_BGR2RGB))
plt.show()
plt.imshow(cv2.cvtColor(255 - barbara, cv2.COLOR_BGR2RGB))
```

```python
#A linear transform can also adjust the brightness and contrast of an image.
#The linear transformation model used is a * I[i, j] + b where a is constrast control and b is the brightness control.
#Insert of array operations, we can use a function for this.
alpha = 2
beta = 0

new_image = cv2.convertScaleAbs(barbara, alpha=alpha, beta=beta)
#The reason we use this function is it calculates the absolute values hence the intensity values fall between 0-255
plt.imshow(cv2.cvtColor(new_image, cv2.COLOR_BGR2RGB))
```

```python
#Histogram equalization is an algorithm that optimizes the contrast.
new_image = cv2.equalizeHist(barbara)
plt.imshow(new_image, cmap="gray")
#Basically the function equalizeHist flattens the histogram and bring all bins' frequences on the same level.
```
