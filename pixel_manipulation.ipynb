```python
from PIL import Image, ImageDraw, ImageFont
from matplotlib import pyplot as plt
import cv2
```

```python
#cropping an image
lenna = Image.open("lenna.png")

plt.imshow(lenna)
```

```python
left = 100
right = 400
upper = 70
lower = 400
lenna_crop = lenna.crop((left, upper, right, lower)) # This will crop the image to the given dimensions

plt.imshow(lenna_crop)
```

```python
#To the paste the elements of one image to another image in a numpy format is 
#I[rowStart:rowEnd, colStart:colEnd, :] = toCopy[rowStart:rowEnd, colStart:colEnd, :]
```

```python
lenna_draw = lenna.copy()
lenna_fn = ImageDraw.Draw(im=lenna_draw) # Now we can draw on the image lenna_draw
print(type(lenna_draw), type(lenna_fn))
plt.imshow(lenna_draw)
```

```python
shape = [left, upper, right, lower]
lenna_fn.rectangle(xy=shape, fill="red") # xy are the bounding box and fill is the color to fill with.
plt.imshow(lenna_draw)
```

```python
baboon = Image.open("baboon.png")

plt.imshow(baboon)
```

```python
baboon_crop = baboon.crop((140, 40, 365, 100))
plt.imshow(baboon_crop)
```

```python
baboon_resize = baboon_crop.resize(size=[110,25])
plt.imshow(baboon_resize)
```

```python
lenna_copy = lenna.copy()
lenna_copy.paste(baboon_resize, box=[245, 260])
plt.imshow(lenna_copy)
```

```python
plt.imshow(lenna_copy.crop(box=[left, upper, right, lower]))
```

```python
cv_lenna = cv2.imread("lenna.png")
start_point, end_point = (left, upper), (right, lower)
# Will draw a rectangle boundary between start point and end point
cv2.rectangle(cv_lenna, pt1=start_point, pt2=end_point, color=(0, 255, 0), thickness=3)

plt.imshow(cv2.cvtColor(cv_lenna, cv2.COLOR_BGR2RGB))
```

```python
cv2.putText(img=cv_lenna, text="Stuff", org=(10, 500), color=(255, 255, 255), fontFace=4, fontScale=5, thickness=2)
#img is the image to draw on, text is the text. org is the bottom left corner to display the text on.
#color is the RGB values, fontFace is the font type, fontScale is the size of the font.

plt.imshow(cv2.cvtColor(cv_lenna, cv2.COLOR_BGR2RGB))
```
