```python
image = cv2.imread("image.jpg") #loads an image

print(type(image)) # A numpy array, Intensity values are 8-bit.
print(image.shape) # Shape of the array (WidthxHeightx3) for colored images.
```

```python
plt.imshow(image) # The colors are off. The reason being the order of each channel is different in openCV.
#OpenCV has R = 0, G = 1, B = 2, whereas PIL has B = 0, G = 1, R = 2. PIL is RGB and OpenCV is BGR.
```

```python
new_image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB) # convert the image from BGR TO RGB(changes the colorspace).
plt.imshow(new_image)
```

```python
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY) # convert the image to grayscale

plt.imshow(image_gray, cmap='gray')
cv2.imwrite("image_gray.png", image_gray) # saving the image
```

```python
#To load grayscale images in openCV, we must specify the colorspace.
im_gray = cv2.imread("barbara.png", cv2.IMREAD_GRAYSCALE)

print(im_gray.shape)
plt.imshow(im_gray, cmap="gray")
```

```python
baboon = cv2.imread("baboon.png")

blue, green, red = baboon[:, :, 0], baboon[:, :, 1], baboon[:, :, 2] #Slicing the numpy array.

figure, axis = plt.subplots(1, 3)

axis[0].set_title("Blue")
axis[0].imshow(blue, cmap="gray")
axis[1].set_title("Green")
axis[1].imshow(green, cmap="gray")
axis[2].set_title("Red")
axis[2].imshow(red, cmap="gray")
```
