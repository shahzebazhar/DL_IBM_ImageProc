```python
from PIL import Image
import cv2
import numpy as np
from PIL import ImageOps
from matplotlib import pyplot as plt
```

```python
baboon = cv2.imread("baboon.png") # this line and the line below are equivalent
baboon = np.array(Image.open("baboon.png"))
```

```python
baboon = cv2.imread("baboon.png")
print(id(baboon)) #object's memory address
```

```python
B = baboon.copy() # Creates a copy, normally setting both variables equal to each other creates a reference.
print(id(B), id(B) == id(baboon))
```

```python
#Flipping an image: Changing an image's orientation
#We can flip images by changing or swapping the index values of the intensities.
lenna = Image.open("lenna.png")
lenna_flip = ImageOps.flip(lenna)

plt.imshow(lenna)
```

```python
plt.imshow(lenna_flip)
```

```python
#Mirror function mirrors the image.
lenna_mirror = ImageOps.mirror(lenna)
plt.imshow(lenna_mirror)
```

```python
#The image object also has a builtin method called transpose, that flips the image.
lenna_flip = lenna.transpose(Image.FLIP_TOP_BOTTOM) # upside down.
plt.imshow(lenna_flip)
```

```python
#Flipping images using openCV
lenna = cv2.imread("lenna.png")
im_flip = cv2.flip(lenna, 0) # 0 is the flipcode that flips on the y axis or makes the image upside down.
im_flip = cv2.rotate(lenna, cv2.ROTATE_90_CLOCKWISE) #rotates the image

plt.imshow(cv2.cvtColor(im_flip, cv2.COLOR_BGR2RGB))
```
